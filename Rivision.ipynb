```python
class IsPalindrome:
    def isPalindrone(self, num):
        temp = num
        x = 0
        while temp > 0:
            rem = temp % 10
            x = (x*10) + rem
            temp//=10
        return num == x

pal = IsPalindrome()
pal.isPalindrone(12221)
```

# Fibonachi

```python
class Fibonachi:
    def fibonachi(self, n):
        if n==0 or n==1:
            return 1
        return self.fibonachi(n-1) + self.fibonachi(n-2)

fib = Fibonachi()
fib.fibonachi(4)
```

# IsPowerOfTwo

```python
class IsPowerOfTwo:
    def isPowerOfTwo(self, n) -> bool:
        if n<=0:
            return False
        while n%2==0:
            n//=2
        return n==1

power = IsPowerOfTwo()
power.isPowerOfTwo(4)
```

# factorial

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact

factorial(5)
```

# Find Maximum in List

```python
def max_find(lst):
    max_num = lst[0]
    for num in lst:
        if num > max_num:
            max_num = num
    return max_num

lst = [7,4,6,3,9,3,8,1]
max_find(lst)
```

# Count Frequency of Characters

```python
def char_frequency(st):
    freq = {}
    for ch in st:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    return freq
char_frequency("nama")
```

# Prime Number Check

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i==0:
            return False
    return True

is_prime(27)
```

# count words from file

```python
def count_words(filename):
    with open(filename, "r")as file:
        text = file.read()
        words = text.split()
        return len(words)

count_words("sample.txt")
```

# Longest Word in Sentence

```python
def longest_word(sen):
    words = sen.split()
    return max(words, key=len)

print(longest_word("I am RUshi learning"))
```

# Find Missing Number (0 to n)

```python
def missing_num(arr):
    n = len(arr)
    total = n * (n + 1) // 2
    return total - sum(arr)

print(missing_num([0,1,2,4]))
```

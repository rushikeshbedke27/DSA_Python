# 1. FizzBuzz

```python
from typing import List

class FizzBuzz:
    def fizzBuzz(self, n: int) -> List[str]:
        ans = []
        for i in range(1, n+1):
            if i%3==0 and i%5==0:
                ans.append("FizzBuzz")
            elif i%3==0:
                ans.append("Fizz")
            elif i%5==0:
                ans.append("Buzz")
            else:
                ans.append(str(i))
        return ans

fz = FizzBuzz()
print(fz.fizzBuzz(15))
```

## 1. FizzBuzz

```python
from typing import List

class FizzBuzz:
    def fizzBuzz(self, n: int) -> List[str]:
        ans = []
        for i in range(1, n+1):
            if i%3==0 and i%5==0:
                ans.append("FizzBuzz")
            elif i%3==0:
                ans.append("Fizz")
            elif i%5==0:
                ans.append("Buzz")
            else:
                ans.append(str(i))
        return ans

fz = FizzBuzz()
print(fz.fizzBuzz(15))
```

## 2. find Odd nums betn specific range

```python
class CountOdds:
    def countOdds(self, low: int, high: int) -> int:
        return (high+1)//2 - (low//2)

odd = CountOdds()
odd.countOdds(3,13)
```

## 3. How many numbers are the smaller than the current num.

```python
from typing import List
class SmallerNumThanCurrent:
    def smallerNumThanCurrent(self, nums: List[int]) -> List[int]:
        ans = []
        for i in nums:
            c = 0
            for j in nums:
                if j < i:
                    c+=1
            ans.append(c)
        return ans

l1 = [8,1,2,2,3]
sntc = SmallerNumThanCurrent()
sntc.smallerNumThanCurrent(l1)
```

## 4 Count the digits that divide a num

```python
class CountDigits:
    def countDigits(self, num: int) -> int:
        temp = num
        c = 0
        while temp > 0:
            rem = temp % 10
            if num % rem == 0:
                c += 1
            temp //= 10
        return c

num1 = 121
num2 = 1248

cd = CountDigits()
cd.countDigits(num2)
```

## 5. Checking number is palindrom or not

```python
class IsPalindrome:
    def isPalindrome(self, num: int) -> bool:
        temp = num
        x = 0
        while temp > 0:
            rem = temp % 10
            x = (x * 10) + rem
            temp //= 10
        return num == x

pal = IsPalindrome()
pal.isPalindrome(12221)
```

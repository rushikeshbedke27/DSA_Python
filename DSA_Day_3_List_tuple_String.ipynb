# 1. Running sum og 1d array

```python
class RunningSum:
    def runningSum(self, nums):
        n = len(nums)
        ans = []
        ans.append(nums[0])
        for i in range(1,n):
            x = ans[i-1] +nums[i]
            ans.append(x)
        return ans


nums = [1,2,3,4]
rs = RunningSum()
print(rs.runningSum(nums))
```

# Fibonachi serires

```python
# Recursive Tree

class Fiboachi:
    def fibonachi(self, n: int) -> int:
        if n==0 or n==1:
            return n
        return self.fibonachi(n-1) + self.fibonachi(n-2) 

    # N-th Tribonachi number
    def tribonachi(self, n: int) -> int:
        if n==0 or n==1:
            return n
        if n==2:
            return 1
        return self.tribonachi(n-1) + self.tribonachi(n-2) + self.tribonachi(n-3)

fi = Fiboachi()
print(fi.fibonachi(7))
print(fi.tribonachi(4))
```

# Power of two

```python
class IsPowerOfTwo:
    # Method 1
    def isPowerOfTwoM1(self, n: int) -> bool:
        if n<=0:
            False
        while n%2==0:
            n//=2
            return n==1
    
    # Method 1
    def isPowerOfTwoM2(self, n: int) -> bool:
        if n<=0:
            return False
        if n==1:
            return True
        if n%2!=0:
            return False
        
        return self.isPowerOfTwoM2(n//2)

pow = IsPowerOfTwo()
print("M1 - ",pow.isPowerOfTwoM1(20))
print("M1 - ",pow.isPowerOfTwoM2(20))
```

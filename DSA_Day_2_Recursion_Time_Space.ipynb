# Fibonachi serires

```python
# Recursive Tree

class Fiboachi:
    def fibonachi(self, n: int) -> int:
        if n==0 or n==1:
            return n
        return self.fibonachi(n-1) + self.fibonachi(n-2) 

    # N-th Tribonachi number
    def tribonachi(self, n: int) -> int:
        if n==0 or n==1:
            return n
        if n==2:
            return 1
        return self.tribonachi(n-1) + self.tribonachi(n-2) + self.tribonachi(n-3)

fi = Fiboachi()
print(fi.fibonachi(7))
print(fi.tribonachi(4))
```

# Power of two

```python
class IsPowerOfTwo:
    # Method 1
    def isPowerOfTwoM1(self, n: int) -> bool:
        if n<=0:
            False
        while n%2==0:
            n//=2
            return n==1
    
    # Method 1
    def isPowerOfTwoM2(self, n: int) -> bool:
        if n<=0:
            return False
        if n==1:
            return True
        if n%2!=0:
            return False
        
        return self.isPowerOfTwoM2(n//2)

pow = IsPowerOfTwo()
print("M1 - ",pow.isPowerOfTwoM1(20))
print("M1 - ",pow.isPowerOfTwoM2(20))
```

## GCD - Greatest Common Factor
## HCF - Highest Commom Factor
## LCM - Least Commom Factor

```python
# Using Recursion

class GCDAndLCDM:
    def gcd(self, a:int, b:int) -> int:
        if b==0:
            return a
        return self.gcd(b, a%b)

    def lcm(self, a:int, b:int) -> int:
        return a*b//self.gcd(a,b)

gl = GCDAndLCDM()
print(gl.gcd(15,50))
print(gl.lcm(15, 50))
```

# pow(x, n)

```python
class PowerX_N:
    def findPow(self, x:float, n:int) -> float:
        # base case
        if n==0:
            return 1
        # resursive
        a = self.findPow(x, n//2)
        if n%2==0:
            return a*a
        else:
            return a*a*x
    
    # But the given power is -ve(2^-2) then we are using below method
    def myPow(self, x:float, n:int) -> float:
        if n>=0:
            return self.findPow(x,n)
        else:
            return 1/self.findPow(x,n*(-1))


pxn = PowerX_N()
print(pxn.myPow(2,10))
print(pxn.myPow(2,-2))
```
